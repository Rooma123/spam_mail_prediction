```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
mail_data=pd.read_csv('/content/mail_data.csv')
```

```python
mail_data.shape
```

```python
mail_data.head()
```

label encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
labels=label_encoder.fit_transform(mail_data.Category)
mail_data['Category']=labels
```

```python
mail_data.head()
```

ham =>0

spam =>1

```python
#separate the data into text and labels
x=mail_data['Message']
y=mail_data['Category']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2,stratify=y)
```

```python
x.shape,x_train.shape,x_test.shape
```

Feature Extraction

```python
#transform the text data to feature vectors that can be used as input to logistic model
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

x_train_features=feature_extraction.fit_transform(x_train)
x_test_features=feature_extraction.transform(x_test)

#convert y_train and y_test to integers
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(x_train)
```

```python
print(x_train_features)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train_features,y_train)
```

```python
x_train_prediction=model.predict(x_train_features)
x_train_accuracy=accuracy_score(x_train_prediction,y_train)
print('the accuracy of train data :', x_train_accuracy)
```

```python
x_test_prediction=model.predict(x_test_features)
x_test_accuracy=accuracy_score(x_test_prediction,y_test)
print('the accuracy of test data :', x_test_accuracy)
```
